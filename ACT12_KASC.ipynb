```python
import random
import yaml
import pandas as pd
import json
import csv

def crear_diccionario_A(ids_existentes):
    ID_a = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
    NOM_H_a = ["Alejandro", "Pedro", "Ricardo", "Angel", "Fernando", "Adrian", "Joel", "Enrique", "Yael", "Cristiano"]
    NOM_M_a = ["Alejandra", "Yessica", "Dianna", "Gabriela", "Fernanda", "Erika", "Danna", "Monica", "Abril", "Ester"]
    AP_a = ["Gomez", "Sanchez", "Espinoza", "Ortega", "Solorzano", "Cordero", "Fuentes", "Ruiz", "Bolaños", "Yepiz"]
    AM_a = ["Lopez", "Dominguez", "Fernandez", "Valenzuela", "Coronado", "Bonilla", "Bernal", "Bravo", "Arreola", "Zayas"]
    ED_a = [20, 25, 40, 33, 46, 29, 48, 67, 19, 55]

    def asignar_sexo(nombre):
        if nombre in NOM_H_a:
            return "Hombre"
        elif nombre in NOM_M_a:
            return "Mujer"

    ID_a_disponibles = [id for id in ID_a if id not in ids_existentes]
    if not ID_a_disponibles:
        print("No hay IDs disponibles para agregar automáticamente.")
        return None

    ID_a_random = random.choice(ID_a_disponibles)
    NOM_a_random = random.choice(NOM_H_a + NOM_M_a)
    AP_a_random = random.choice(AP_a)
    AM_a_random = random.choice(AM_a)
    ED_a_random = random.choice(ED_a)
    SEX_a_random = asignar_sexo(NOM_a_random)

    empleado = {"ID": ID_a_random, "Nombre": NOM_a_random, "Apellido Paterno": AP_a_random, "Apellido Materno": AM_a_random, "Edad": ED_a_random, "Sexo": SEX_a_random}
    return empleado

def crear_diccionario():
    while True:
        id = int(input("Ingrese su id: "))
        if any(emp["ID"] == id for emp in lista_empleados):
            print("El ID ya existe. Ingrese un ID único.")
        else:
            break
    nom = input("Ingrese su nombre (Uno o Dos): ")
    ap = input("Ingrese su apellido paterno: ")
    am = input("Ingrese su apellido materno: ")
    ed = input("Ingrese su edad: ")
    sex = input("Ingrese su sexo: ")
    empleado = {"ID": id, "Nombre": nom, "Apellido Paterno": ap, "Apellido Materno": am, "Edad": ed, "Sexo": sex}
    return empleado

def imprimir_lista(empleados):
    if empleados:
        df = pd.DataFrame(empleados)
        print(df.to_markdown(index=False))
    else:
        print("La lista de empleados está vacía.")

def buscar_id(empleados, id_buscar):
    for empleado in empleados:
        if empleado["ID"] == id_buscar:
            print("\nEmpleado encontrado:")
            print(yaml.dump(empleado))
            return
    print("Empleado no encontrado.")

def buscar_appat(empleados, appat_buscar):
    coincidencias = [emp for emp in empleados if emp["Apellido Paterno"].lower() == appat_buscar.lower()]
    if coincidencias:
        print(f"\nEmpleados encontrados con apellido paterno '{appat_buscar}':")
        for emp in coincidencias:
            print(yaml.dump(emp))
    else:
        print("No se encontraron coincidencias.")

def ordenar_lista(empleados):
    if empleados != sorted(empleados, key=lambda x: x["ID"]):
        return sorted(empleados, key=lambda x: x["ID"])
    else:
        print("La lista ya está ordenada por ID.")
        return empleados

def generar_archivo(empleado, formato):
    nombre_archivo = input("Ingrese el nombre del archivo (sin extensión): ")
    if formato == "excel":
        df = pd.DataFrame([empleado])
        df.to_excel(f"{nombre_archivo}.xlsx", index=False)
    elif formato == "txt":
        with open(f"{nombre_archivo}.txt", 'w') as file:
            file.write(json.dumps(empleado, indent=4))
    elif formato == "csv":
        df = pd.DataFrame([empleado])
        df.to_csv(f"{nombre_archivo}.csv", index=False)
    elif formato == "markdown":
        with open(f"{nombre_archivo}.md", 'w') as file:
            file.write(f"| ID | Nombre | Apellido Paterno | Apellido Materno | Edad | Sexo |\n")
            file.write(f"|----|--------|------------------|------------------|------|------|\n")
            file.write(f"| {empleado['ID']} | {empleado['Nombre']} | {empleado['Apellido Paterno']} | {empleado['Apellido Materno']} | {empleado['Edad']} | {empleado['Sexo']} |\n")
    else:
        print("Formato no válido.")

def cargar_archivo(formato):
    nombre_archivo = input("Ingrese el nombre del archivo (sin extensión): ")
    if formato == "excel":
        df = pd.read_excel(f"{nombre_archivo}.xlsx")
        return df.to_dict('records')
    elif formato == "txt":
        with open(f"{nombre_archivo}.txt", 'r') as file:
            contenido = file.read()
            return [json.loads(contenido)]
    elif formato == "csv":
        df = pd.read_csv(f"{nombre_archivo}.csv")
        return df.to_dict('records')
    elif formato == "markdown":
        with open(f"{nombre_archivo}.md", 'r') as file:
            contenido = file.readlines()
            empleado = {
                "ID": int(contenido[2].split('|')[1].strip()),
                "Nombre": contenido[2].split('|')[2].strip(),
                "Apellido Paterno": contenido[2].split('|')[3].strip(),
                "Apellido Materno": contenido[2].split('|')[4].strip(),
                "Edad": int(contenido[2].split('|')[5].strip()),
                "Sexo": contenido[2].split('|')[6].strip()
            }
            return [empleado]
    else:
        print("Formato no válido.")
        return []

def borrar_toda_lista():
    confirmacion = input("¿Está seguro que desea borrar toda la lista de empleados? (s/n): ")
    if confirmacion.lower() == "s":
        lista_empleados.clear()
        print("Lista de empleados borrada")
    else:
        print("Operación cancelada.")

def main():
    global lista_empleados
    lista_empleados = []

    while True:
        print("\n--------- MENU ----------\n1.- Agregar (automático 10)\n2.- Eliminar {ID}\n3.- Imprimir lista (tabla)\n4.- Buscar {ID}\n5.- Buscar {appat} todas las coincidencias\n6.- Ordenar {ID}\n7.- Generar archivo {ID}\n8.- Cargar archivo {ID}\n9.- Imprimir archivo {ID}\n10.- Borrar Toda la lista {ID}\n0.- SALIR")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            for _ in range(10):
                empleado = crear_diccionario_A([emp["ID"] for emp in lista_empleados])
                if empleado:
                    lista_empleados.append(empleado)
            print("10 empleados agregados automáticamente")
        elif opcion == "2":
            if not lista_empleados:
                print("La lista de empleados está vacía.")
            else:
                id_eliminar = int(input("Ingrese el ID del empleado a eliminar: "))
                empleado = next((emp for emp in lista_empleados if emp["ID"] == id_eliminar), None)
                if empleado:
                    confirmacion = input(f"¿Está seguro que desea eliminar al empleado con ID {id_eliminar}? (s/n): ")
                    if confirmacion.lower() == "s":
                        lista_empleados.remove(empleado)
                        print("Empleado eliminado.")
                    else:
                        print("Operación cancelada.")
                else:
                    print("No se encontró ningún empleado con ese ID.")
        elif opcion == "3":
            imprimir_lista(lista_empleados)
        elif opcion == "4":
            if not lista_empleados:
                print("La lista de empleados está vacía.")
            else:
                id_buscar = int(input("Ingrese el ID a buscar: "))
                buscar_id(lista_empleados, id_buscar)
        elif opcion == "5":
            if not lista_empleados:
                print("La lista de empleados está vacía.")
            else:
                appat_buscar = input("Ingrese el Apellido Paterno a buscar: ")
                buscar_appat(lista_empleados, appat_buscar)
        elif opcion == "6":
            if not lista_empleados:
                print("La lista de empleados está vacía.")
            else:
                lista_empleados = ordenar_lista(lista_empleados)
                if lista_empleados != sorted(lista_empleados, key=lambda x: x["ID"]):
                    lista_empleados = sorted(lista_empleados, key=lambda x: x["ID"])
                    print("Lista de empleados ordenada por ID.")
        elif opcion == "7":
            if not lista_empleados:
                print("La lista de empleados está vacía.")
            else:
                id_buscar = int(input("Ingrese el ID del empleado para generar archivo: "))
                empleado = next((emp for emp in lista_empleados if emp["ID"] == id_buscar), None)
                if empleado:
                    print("Seleccione el formato del archivo:\na) Excel\nb) Txt\nc) Csv\nd) Markdown")
                    formato = input("Seleccione una opción: ").lower()
                    if formato in ['a', 'b', 'c', 'd']:
                        formato_dict = {'a': 'excel', 'b': 'txt', 'c': 'csv', 'd': 'markdown'}
                        generar_archivo(empleado, formato_dict[formato])
                    else:
                        print("Opción no válida.")
                else:
                    print("Empleado no encontrado.")
        elif opcion == "8":
            print("Seleccione el formato del archivo:\na) Excel\nb) Txt\nc) Csv\nd) Markdown")
            formato = input("Seleccione una opción: ").lower()
            if formato in ['a', 'b', 'c', 'd']:
                formato_dict = {'a': 'excel', 'b': 'txt', 'c': 'csv', 'd': 'markdown'}
                empleados_cargados = cargar_archivo(formato_dict[formato])
                if empleados_cargados:
                    for emp in empleados_cargados:
                        if any(e["ID"] == emp["ID"] for e in lista_empleados):
                            print(f"El empleado con ID {emp['ID']} ya existe en la lista.")
                        else:
                            lista_empleados.append(emp)
                    print("Archivo cargado exitosamente.")
            else:
                print("Opción no válida.")
        elif opcion == "9":
            print("Seleccione el formato del archivo:\na) Excel\nb) Txt\nc) Csv\nd) Markdown")
            formato = input("Seleccione una opción: ").lower()
            if formato in ['a', 'b', 'c', 'd']:
                formato_dict = {'a': 'excel', 'b': 'txt', 'c': 'csv', 'd': 'markdown'}
                empleados_cargados = cargar_archivo(formato_dict[formato])
                if empleados_cargados:
                    imprimir_lista(empleados_cargados)
            else:
                print("Opción no válida.")
        elif opcion == "10":
            borrar_toda_lista()
        elif opcion == "0":
            print("Saliendo...")
            break
        else:
            print("Opción no válida, ingrese una opción válida.")

if __name__ == "__main__":
    main()
```
