# PRACTICA #10

REALIZA UN PROGRAMA QUE UTILICE UNA LISTA DE DICCIONARIOS

MENU

1.- Agregar (automatico)

2.- Agregar (manual)

3.- Imprimir lista

4.- Buscar {ID}

5.- Ordenar

6.- Eliminar {ID}

7.- Borrar Toda la lista {ID}

0.- SALIR

Preguntar siempre si esta seguro eliminar, No ordenar si ya esta Ordenado, decir si lista vacía, no existe o mostrar si estala búsqueda, No se permiten ID repetidos


```python
import random
import yaml
import pandas as pd
import json

def crear_diccionario_A():
    # Automático
    ID_a = {1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000}
    NOM_H_a = {"Alejandro", "Pedro", "Ricardo", "Angel", "Fernando", "Adrian", "Joel", "Enrique"}
    NOM_M_a = {"Alejandra", "Yessica", "Dianna", "Gabriela", "Fernanda", "Erika", "Danna", "Monica"}
    AP_a = {"Gomez", "Sanchez", "Espinoza", "Ortega", "Solorzano", "Cordero", "Fuentes", "Ruiz"}
    AM_a = {"Lopez", "Dominguez", "Fernandez", "Valenzuela", "Coronado", "Bonilla", "Bernal", "Bravo"}
    ED_a = {20, 25, 40, 33, 46, 29, 48, 67}

    def asignar_sexo(nombre):
        if nombre in NOM_H_a:
            return "Hombre"
        elif nombre in NOM_M_a:
            return "Mujer"

    ID_a_random = random.choice(tuple(ID_a))
    NOM_a_random = random.choice(tuple(NOM_H_a.union(NOM_M_a)))
    AP_a_random = random.choice(tuple(AP_a))
    AM_a_random = random.choice(tuple(AM_a))
    ED_a_random = random.choice(tuple(ED_a))
    SEX_a_random = asignar_sexo(NOM_a_random)

    empleado = {"ID": ID_a_random, "Nombre": NOM_a_random, "Apellido Paterno": AP_a_random, "Apellido Materno": AM_a_random, "Edad": ED_a_random, "Sexo": SEX_a_random}
    return empleado

def crear_diccionario():
    # Manual
    lista = []
    id = int(input("Ingrese su id: "))
    lista.append(id)
    nom = input("Ingrese su nombre (Uno o Dos): ")
    lista.append(nom)
    ap = input("Ingrese su apellido paterno: ")
    lista.append(ap)
    am = input("Ingrese su apellido materno: ")
    lista.append(am)
    ed = input("Ingrese su edad: ")
    lista.append(ed)
    sex = input("Ingrese su sexo: ")
    lista.append(sex)
    empleado = {"ID": id, "Nombre": nom, "Apellido Paterno": ap, "Apellido Materno": am, "Edad": ed, "Sexo": sex}
    return empleado

def print_yaml(data):
    print(yaml.dump(data))

def print_json(data):
    print(json.dumps(data, indent=4))

def print_dataframe(data, output_format):
    if output_format == "pandas":
        df = pd.DataFrame(data)
        print(df)
    elif output_format == "yaml":
        print_yaml(data)
    elif output_format == "json":
        print_json(data)
    else:
        print("Formato de salida no válido.")

def ordenar_lista(lista, criterio):
    if criterio.lower() == "id":
        return sorted(lista, key=lambda x: int(x[criterio]))
    else:
        return sorted(lista, key=lambda x: x[criterio].lower())

def main():
    lista_empleados = []

    while True:
        print("\n--------- MENU ----------\n1.- Agregar (Automático)\n2.- Agregar (Manual)\n3.- Imprimir lista\n4.- Buscar (ID)\n5.- Ordenar\n6.- Eliminar (ID)\n7.- Borrar Toda la lista\n0.- SALIR")
        opcion = input("Seleccione una opcion: ")

        if opcion == "1":
            empleado = crear_diccionario_A()
            lista_empleados.append(empleado)
            print("Empleado agregado automaticamente")
        elif opcion == "2":
            empleado = crear_diccionario()
            lista_empleados.append(empleado)
            print("Empleado agregado manualmente")
        elif opcion == "3":
            if lista_empleados:
                print("\n¿Como desea imprimir la lista de empleados?")
                print("1. DataFrame de Pandas")
                print("2. YAML")
                print("3. JSON")
                opcion_imprimir = input("Seleccione una opcion: ")
                if opcion_imprimir == "1":
                    print_dataframe(lista_empleados, "pandas")
                elif opcion_imprimir == "2":
                    for empleado in lista_empleados:
                        print_dataframe(empleado, "yaml")
                elif opcion_imprimir == "3":
                    for empleado in lista_empleados:
                        print_dataframe(empleado, "json")
                else:
                    print("Opcion no valida.")
            else:
                print("La lista de empleados esta vacia.")
        elif opcion == "4":
            id_buscar = int(input("Ingrese el ID a buscar: "))
            encontrado = False
            for empleado in lista_empleados:
                if empleado["ID"] == id_buscar:
                    print("\nEmpleado encontrado:")
                    print_yaml(empleado)
                    encontrado = True
                    break
            if not encontrado:
                print("Empleado no encontrado.")
        elif opcion == "5":
            criterio = input("Ingrese el criterio de ordenamiento (ID, Nombre, Edad, Sexo): ").capitalize()
            if criterio in ["ID", "Nombre", "Edad", "Sexo"]:
                lista_empleados = ordenar_lista(lista_empleados, criterio)
                print("Lista de empleados ordenada por", criterio)
            else:
                print("Criterio de ordenamiento no válido.")
        elif opcion == "6":
            id_eliminar = int(input("Ingrese el ID del empleado a eliminar: "))
            eliminado = False
            for empleado in lista_empleados:
                if empleado["ID"] == id_eliminar:
                    lista_empleados.remove(empleado)
                    print("Empleado eliminado.")
                    eliminado = True
                    break
            if not eliminado:
                print("No se encontro ningún empleado con ese ID.")
        elif opcion == "7":
            confirmacion = input("¿Esta seguro que desea borrar toda la lista de empleados? (s/n): ")
            if confirmacion.lower() == "s":
                lista_empleados.clear()
                print("Lista de empleados borrada")
            else:
                print("Operacion cancelada.")
        elif opcion == "0":
            print("Saliendo...")
            break
        else:
            print("Opcion no valida, ingrese una opción valida.")

if __name__ == "__main__":
    main()

```
